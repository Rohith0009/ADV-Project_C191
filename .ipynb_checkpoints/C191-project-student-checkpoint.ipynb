```python
#Covide Data 
import  numpy as np
import pandas as pd
from matplotlib import pyplot as plt


dataframe = pd.read_csv('covid19.csv')
df = dataframe.dropna()
sorted_df = df.sort_values(by=['total_cases'])
cases = sorted_df['total_cases'].head(5)
country = sorted_df['country'].head(5)

plt.xlabel("Country")
plt.xticks(rotation="vertical")
plt.ylabel("Cases")

label = country
value = cases

plt.bar(label, value, width=0.5, color=("palegreen", "paleturquoise", "red", "yellow", "black"))
```

```python
#Task 1 
#Sort the data as per total number of cases
sorted_df = df.sort_values(by=['total_deaths'])
deaths = sorted_df['total_deaths'].tail(5)
country = sorted_df['country'].tail(5)

plt.xlabel("Country")
plt.xticks(rotation="vertical")
plt.ylabel("Cases")

label = country
value = deaths

plt.bar(label, value, width=0.5, color=("palegreen", "paleturquoise", "red", "yellow", "black"))
```
